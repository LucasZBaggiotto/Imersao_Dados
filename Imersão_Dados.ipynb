```python
#desafio 01 -- 
import seaborn as sns
sns.set()
plt.figure(figsize=(15,15))

ax = sns.countplot(x = 'composto' , data = dados.query('composto in @cod_compostos') , order=cod_compostos, palette='dark:pink_r')

ax.set_title('Top 5 compostos' , style = 'italic' , fontweight='bold', fontsize = '33')
plt.ylabel('Quantidade')
plt.xlabel('Composto')
plt.show()
```

```python
# fazendo um gráfico com os 10 compostos mais utilizados
plt.figure(figsize=(20,20))
com_drogas = dados.query('tratamento == "com_droga"')
dados_composto50 = com_drogas.composto.value_counts().head(50).index #os 50 mais na ordem
dados_comdroga50 = dados.query('composto in @dados_composto50') 
ax = sns.countplot(y = 'composto', data= dados_comdroga50, palette ='dark:salmon', order=dados_composto50)
ax.set_title('Top 50 compostos', fontsize =30, style = 'italic')
plt.xlabel('Quantidade do Composto', fontsize = 20)
plt.ylabel('Composto')
```

```python
#brincando um pouco com a HUE
plt.figure(figsize=(20,20))

com_drogas = dados.query('tratamento == "com_droga"')
dados_composto10 = com_drogas.composto.value_counts().head(10).index 
dados_comdroga10 = dados.query('composto in @dados_composto10') 
ax = sns.countplot(x = 'composto', data= dados_comdroga10, palette ='dark:salmon', order=dados_composto10, hue ='tempo')
ax.set_title('Top 10 compostos', fontsize =30, style = 'italic')
plt.xlabel('Composto', fontsize = 20)
plt.ylabel('Quantidade do Composto')
```
